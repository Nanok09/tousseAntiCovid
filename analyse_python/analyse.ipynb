```python
df2['Var SPO2']=0
```

```python
df2['score']=0
```

```python
df2
```

```python
for i in range(1,len(df2)):
    for k in range(i):
        if df2.loc[k,'ID']=df2.loc[i,'ID']:
            if (df2.loc[i,'Date']-df2.loc[k,'Date'])>0:
                df2.loc[i,'Var SPO2']=60*abs((df2.loc[i,'SPO2']-df2.loc[k,'SPO2'])/(df2.loc[i,'Date']-df2.loc[k,'Date']))
            else:
                df2.loc[i,'Var SPO2']=60*abs((df2.loc[i,'SPO2']-df2.loc[k,'SPO2'])/(df2.loc[i,'Date']+86400-df2.loc[k,'Date']))
        
```

```python
for i in range(len(df2)):
    if df2.loc[i,'Risque']=='oui':
        df2.loc[i,'score']=3
        if df2.loc[i,'SPO2']<=91 or df2.loc[i,'FR']>30 or df2.loc[i,'Var SPO2']>1:
            df2.loc[i,'score']=4
    else:
        df2.loc[i,'score']=1  
        if df2.loc[i,'T°']>38.5 or df2.loc[i,'FR']>20 or df2.loc[i,'PAS']<100:
            df2.loc[i,'score']=2
            if df2.loc[i,'SPO2']<96 or df2.loc[i,'FR']>24 or df2.loc[i,'PAS']<90 or df2.loc[i,'T°']>40 or df2.loc[i,'Var SPO2']>0.5:
                df2.loc[i,'score']=3
                if df2.loc[i,'SPO2']<91 or df2.loc[i,'FR']>30 or df2.loc[i,'Var SPO2']>1:
                    df2.loc[i,'score']=4
```
